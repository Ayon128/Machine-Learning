```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
Classes = ["Python","ML","AI","R","DS"]
Class_1 = [10,20,30,40,50]
Class_2 = [40,50,10,30,35]
Class_3=  [50,35,37,40,45]
```

```python
plt.bar(Classes , Class_1)
```

```python
plt.barh(Classes , Class_1)
```

```python
plt.bar(Classes , Class_1,width = 0.5)
```

```python
plt.bar(Classes , Class_1,width = 0.5,align = "edge")
```

```python
plt.bar(Classes , Class_1,width = 0.5,align = "edge",color = "y")
```

```python
plt.bar(Classes , Class_1,width = 0.5,align = "edge",color = "y",edgecolor = "k")
```

```python
plt.bar(Classes , Class_1,width = 0.5,align = "edge",color = "y",edgecolor = "k",linewidth = 5)
```

```python
plt.bar(Classes , Class_1,width = 0.5,align = "edge",color = "y",edgecolor = "k",linewidth = 5,alpha = 0.9)
```

```python
plt.bar(Classes , Class_1,width = 0.5,align = "edge",color = "y",edgecolor = "k",linewidth = 5,alpha = 0.9,linestyle = "--")
```

```python
plt.bar(Classes , Class_1,width = 0.5,align = "edge",color = "y",edgecolor = "k",linewidth = 5,alpha = 0.9,linestyle = "--",label = "Students bar chart")
plt.legend()
```

```python
plt.bar(Classes , Class_1,width = 0.5,align = "edge",color = "y",edgecolor = "k",linewidth = 5,alpha = 0.9,linestyle = "--",label = "Students bar chart",visible =False)
plt.legend()
```

```python
style.use("ggplot")
plt.figure(figsize = (16,9))
plt.bar(Classes , Class_1,width = 0.5,align = "edge",color = "y",edgecolor = "k",linewidth = 5,alpha = 0.9,linestyle = "--",label = "Students bar chart")
```

```python
style.use("ggplot")
plt.figure(figsize = (16,9))
plt.barh(Classes , Class_1,align = "edge",color = "y",edgecolor = "k",linewidth = 5,alpha = 0.9,linestyle = "--",label = "Students bar chart")
```

```python
style.use("ggplot")
plt.figure(figsize = (16,9))
plt.bar(Classes , Class_1,width = 0.5,align = "edge",color = "y",edgecolor = "k",linewidth = 5,alpha = 0.9,linestyle = "--")
plt.title("Bar Chart of IAIP Class")
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("No. of Students",fontsize = 15)
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize = (16,9))
plt.bar(Classes , Class_1,width = 0.5,align = "edge",color = "y",label = "Class_1 Students")
plt.bar(Classes , Class_2,width = 0.5,align = "edge",color = "g",label = "Class_2 Students")
plt.title("Bar Chart of IAIP Class")
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("No. of Students",fontsize = 15)
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize = (16,9))
Classes_index=np.arange(len(Classes))
width= 0.3
plt.bar(Classes_index , Class_1,width,align = "edge",color = "y",label = "Class_1 Students")
plt.bar(Classes_index+width, Class_2,width,align = "edge",color = "g",label = "Class_2 Students")
plt.xticks(Classes_index + width,Classes)
plt.title("Bar Chart of IAIP Class")
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("No. of Students",fontsize = 15)
plt.show()
```
