```python
import matplotlib.pyplot as plt
```

```python
Class = ["Python", "R","AI","ML","Data Sciece"]
Class_Students = [45,35,25,15,10]
```

```python
plt.pie([1])
plt.show()
```

```python
plt.pie(Class_Students)
plt.show()
```

```python
plt.pie(Class_Students,labels = Class)
plt.show()
```

```python
explode = [0,0,0,0,0]
plt.pie(Class_Students,labels = Class, explode = explode)
plt.show()
```

```python
explode = [0,.3,0,0,0]
plt.pie(Class_Students,labels = Class, explode = explode)
plt.show()
```

```python
explode = [0,.3,0,0,0]
colors = ['g','y','m','b','c']
plt.pie(Class_Students,labels = Class, explode = explode,colors = colors)
plt.show()
```

```python
explode = [0,.3,0,0,0]
colors = ['g','y','m','b','c']
plt.pie(Class_Students,labels = Class, explode = explode,colors = colors,autopct = "%.1f%%")
plt.show()
```

```python
explode = [0,.3,0,0,0]
colors = ['g','y','m','b','c']
plt.pie(Class_Students,labels = Class, explode = explode,colors = colors,autopct = "%.1f%%",shadow = True)
plt.show()
```

```python
explode = [0,.3,0,0,0]
colors = ['g','y','m','b','c']
plt.pie(Class_Students,labels = Class, explode = explode,colors = colors,autopct = "%.1f%%",shadow = True,radius = 1.5)
plt.show()
```

```python
explode = [0,.3,0,0,0]
colors = ['g','y','m','b','c']
plt.pie(Class_Students,labels = Class, explode = explode,colors = colors,autopct = "%.1f%%",shadow = True,radius = 1.5,startangle = 90)
plt.show()
```

```python
explode = [0,.3,0,0,0]
colors = ['g','y','m','b','c']
textprops = {"fontsize":15}
plt.pie(Class_Students,labels = Class, explode = explode,colors = colors,autopct = "%.1f%%",shadow = True,radius = 1.5,startangle = 90,textprops = textprops)
plt.show()
```

```python
explode = [0,.3,0,0,0]
colors = ['g','y','m','b','c']
plt.pie(Class_Students,labels = Class, explode = explode,colors = colors,autopct = "%.1f%%",shadow = True,radius = 1.5,startangle = 90,rotatelabels = True,pctdistance=0.6)
plt.show()
```
