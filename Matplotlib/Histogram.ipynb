# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
a=np.random.randint(18,45,100)
b=np.random.randint(15,40,100)
print(a)
print(b)
```

```python
plt.hist(a)
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No.Students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(a,bins)
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No.Students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(a,bins,rwidth = 0.8)
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No.Students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(a,bins,rwidth = 0.8,histtype = "step")
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No.Students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(a,bins,rwidth = 0.8,align = "mid")
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No.Students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(a,bins,rwidth = 0.8,align = "mid",orientation  = "horizontal")
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No.Students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(a,bins,rwidth = 0.8,align = "mid",orientation  = "vertical")
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No.Students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(a,bins,rwidth = 0.8,align = "mid",orientation  = "vertical",color = "m")
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No.Students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(a,bins,rwidth = 0.8,align = "mid",orientation  = "vertical",color = "m",label = "Ml Students")
plt.hist(b,bins,rwidth = 0.8,align = "mid",orientation  = "vertical",color = "y",label = "Py Students")
plt.legend()
plt.legend()
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No.Students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist([a,b],bins,rwidth = 0.8,align = "mid",orientation  = "vertical",color = ["m","y"],label = ["Ml Students","Py Students"])
plt.legend()
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No.Students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.figure(figsize = (16,9))
plt.hist([a,b],bins,rwidth = 0.8,align = "mid",orientation  = "vertical",color = ["m","y"],label = ["Ml Students","Py Students"])
plt.legend()
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No.Students age")
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
bins = [15,20,25,30,35,40,45]
plt.figure(figsize = (16,9))
plt.hist([a,b],bins,rwidth = 0.8,align = "mid",orientation  = "vertical",color = ["m","y"],label = ["Ml Students","Py Students"])
plt.legend()
plt.title("ML students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No.Students age")
plt.show()
```
