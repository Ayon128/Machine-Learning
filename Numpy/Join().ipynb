```python
import pandas as pd
```

```python
df1=pd.DataFrame({'A':[1,2,3],'B':[5,6,7]})
df2=pd.DataFrame({'C' :[8,9,10],'D':[11,12,13]})
display(df1,df2)
```

```python
df1.join(df2)
```

```python
df1=pd.DataFrame({'A':[1,2,3],'B':[5,6,7]},index=['a','b','c'])
df2=pd.DataFrame({'C' :[8,9,10],'D':[11,12,13]})
display(df1,df2)
```

```python
df1.join(df2)#index must be same for join
```

```python
df1=pd.DataFrame({'A':[1,2,3],'B':[5,6,7]},index=['a','b','c'])
df2=pd.DataFrame({'C' :[8,9,10],'D':[11,12,13]},index = ['a','b','c'])
display(df1,df2)
```

```python
df1.join(df2)
```

```python
df1=pd.DataFrame({'A':[1,2,3],'B':[5,6,7]},index=['a','b','c'])
df2=pd.DataFrame({'C' :[9,10],'D':[12,13]},index = ['a','b',])
display(df1,df2) 
```

```python
df1.join(df2)
```

```python
df1.join(df2, how ='outer')
```

```python
df1.join(df2, how ='inner')
```

```python
df1.join(df2, how ='right')
```

```python
df1=pd.DataFrame({'A':[1,2,3],'B':[5,6,7]},index=['a','b','c'])
df2=pd.DataFrame({'A' :[8,9,10],'D':[11,12,13]},index = ['a','b','c'])
display(df1,df2)
```

```python
df1.join(df2,lsuffix = '_1')#lsuffix/rsuffix used when label/coloumn names are same
```

```python
df1.join(df2,rsuffix = '_1')
```
