```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
li = load_iris()
```

```python
df = pd.DataFrame(np.c_[li.data,li.target],columns=[list(li.feature_names)+['targer']])
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test =  train_test_split(X,Y,test_size = 0.2, random_state = 10)
```

```python
from sklearn.ensemble import RandomForestClassifier
rc = RandomForestClassifier(n_estimators=10)
```

```python
rc.fit(X_train,Y_train)
rc.score(X_test,Y_test)
```

```python
rc1 = RandomForestClassifier(n_estimators=50)
rc1.fit(X_train,Y_train)
rc1.score(X_test,Y_test)
```

```python
rc2 = RandomForestClassifier(n_estimators=40)
rc2.fit(X_train,Y_train)
rc2.score(X_test,Y_test)
```
