```python
#Creating DataFrame 
new_df = pd.DataFrame(np.c_[df.data,df.target],columns = [list(df.feature_names)+['target']])
new_df.head()
```

```python
new_df.info()
```

```python
new_df.isnull().sum()
```

```python
X = new_df.iloc[:,0:-1]
Y = new_df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 2020)
```

```python
from sklearn.svm import SVC
svc_rbf = SVC(kernel = 'rbf')
svc_rbf.fit(X_train,Y_train)
```

```python
svc_rbf.score(X_test,Y_test)
```

#Feature Scaling


```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
```

```python
X_train_sc = sc.transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
svc_rbf.fit(X_train_sc,Y_train)
```

```python
svc_rbf.score(X_test_sc,Y_test)
```

# SVC with polynomial kernel

```python
svc_ploy = SVC(kernel = 'poly',degree = 2)
svc_ploy.fit(X_train_sc,Y_train)
svc_ploy.score(X_test_sc,Y_test)
```

# NSVC with Linar kernel

```python
svc_linear = SVC(kernel = 'linear')
svc_linear.fit(X_train_sc,Y_train)
svc_linear.score(X_test_sc,Y_test)
```
