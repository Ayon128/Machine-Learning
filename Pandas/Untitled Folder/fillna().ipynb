```python
import pandas as pd
```

```python
df=pd.read_csv('E:\\Machine_Learning\student_1.csv')
df
```

```python
df.fillna(5)#fillna used to fill the NaN with values 
```

```python
df.fillna(method = 'ffill')#ffil means fill the missing value with (forward fill)previous row value(by default)
```

```python
df.fillna(method = 'bfill')#bfil means fill the missing value with (backward fill)front row  value(by default)
```

```python
df.fillna(method = 'pad')
```

```python
df.fillna(method ='ffill',axis=1)#ffil means fill the missing value with (foreward fill)previous coloumn  value
```

```python
df.fillna(method ='bfill',axis=1)#bfil means fill the missing value with (backward fill)front coloumn  value
```

```python
df.fillna(5,limit=1)#limit=1 means every coloumn which contains NaN replace with value (at least one NaN)
```

```python
df.fillna(5,limit=3)#limit=1 means every coloumn which contains NaN replace with value (Three NaN)
```

```python
df.fillna(method ='ffill',limit=3)#limit=1 means every coloumn which contains NaN replace with previous row value (Three NaN)
```

```python
df.fillna(method = 'ffill',inplace =True)#it doesn't create new DataFrame(mainly update the main DataFrame)
df
```
