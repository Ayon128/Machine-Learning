```python
from sklearn.datasets import load_digits
digit = load_digits()
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(digit.data,digit.target,test_size = 0.3)
```

```python
lr = LogisticRegression(solver = 'liblinear',multi_class= 'ovr',)
lr.fit(X_train,Y_train)
lr.score(X_test,Y_test)
```

```python
svm = SVC(gamma = 'auto')
svm.fit(X_train,Y_train)
svm.score(X_test,Y_test)
```

```python
rf = RandomForestClassifier(n_estimators= 40)
rf.fit(X_train,Y_train)
rf.score(X_test,Y_test)
```

# cross_val_score function

```python
from sklearn.model_selection import cross_val_score
```

# Logistic regression model performance using cross_val_score

```python
cross_val_score(LogisticRegression(solver = 'liblinear',multi_class='ovr'),digit.data,digit.target,cv= 3)
```

# svm model performance using cross_val_score

```python
cross_val_score(SVC(gamma = 'auto'),digit.data,digit.target,cv= 3)
```

#random forest performance using cross_val_score

```python
cross_val_score(RandomForestClassifier(n_estimators=40),digit.data,digit.target,cv= 3)
```

#  cross_val_score uses stratifield kfold by default

# Parameter tunning using k fold cross validation

```python
score1 = cross_val_score(RandomForestClassifier(n_estimators=5),digit.data,digit.target,cv= 10)
np.average(score1)
```

```python
score2 = cross_val_score(RandomForestClassifier(n_estimators=10),digit.data,digit.target,cv= 10)
np.average(score2)
```

```python
score3 = cross_val_score(RandomForestClassifier(n_estimators=20),digit.data,digit.target,cv= 10)
np.average(score3)
```

```python
score4 = cross_val_score(RandomForestClassifier(n_estimators=50),digit.data,digit.target,cv= 10)
np.average(score4)
```

```python
score5 = cross_val_score(RandomForestClassifier(n_estimators=40),digit.data,digit.target,cv= 10)
np.average(score5)
```

### Here we used cross_val_score to fine tune our random forest classifier and figured that having around 50 trees in random forest gives best result.
