### Categorical Variables and One Hot Encoding

```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/5_one_hot_encoding/homeprices.csv")
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
new_df = pd.concat([df,dummies],axis =1)
new_df
```

```python
final = new_df.drop('town',axis ='columns')
final
```

```python
final = final.drop('west windsor',axis =1)
final
```

```python
x = final.drop('price',axis =1)
x
```

```python
y = final.price
y
```

```python
from sklearn.linear_model import LinearRegression
model =LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[3400,0,0]]) # 3400 sqr ft home in west windsor
```

```python
model.predict([[2800,0,1]])# 2800 sqr ft home in robbinsville
```

### Use sklearn OneHotEncoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dl1 = df
dl1.town = le.fit_transform(df.town)
dl1
```

```python
X1 = dl1[['town','area']]
X1
```

```python
X1.values
```

```python
Y1 = dl1[['price']]
Y1
```

```python
Y1.values
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X1 =ct.fit_transform(X1)
X1
```

```python
X1=X1[:,1:]
X1
```

```python
model.fit(X1,Y1)
```

```python
model.predict([[0,1,3400]])       # 3400 sqr ft home in west windsor
```

```python
model.predict([[1,0,2800]]) # 2800 sqr ft home in robbinsville
```
