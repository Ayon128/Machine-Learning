```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/tips.csv")
df.head()
```

```python
dummies = pd.get_dummies(df)
dummies
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse = False)
```

```python
dummies.keys()
```

```python
df.keys()
```

```python
enc = ohe.fit_transform(df[['sex','smoker','day','time']])
enc
```

```python
new_df = pd.DataFrame(enc,columns=['sex_Female', 'sex_Male', 'smoker_No','smoker_Yes', 'day_Fri', 'day_Sat', 'day_Sun', 'day_Thur',
       'time_Dinner', 'time_Lunch'])
new_df
```
